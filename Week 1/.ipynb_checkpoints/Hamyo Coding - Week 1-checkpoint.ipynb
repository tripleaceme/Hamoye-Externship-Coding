```python
#Make sure you import all necessary python Libraries.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Declare a variable for each of the three(3) csv files and read the data.
phone = pd.read_csv('cell_phones.csv')

population = pd.read_csv('population1.csv')

geo_code = pd.read_csv('ddf--entities--geo--country.csv')
```

```python
#Print out the first 5 rows
phone.head()
```

```python
#Print out the first 5 rows
population.head()
```

```python
#Print out the first 5 rows
geo_code.head()
```

```python
#Join the ‘Cell phones’ and ‘population’ dataframe (With an inner join on ‘Country’ and ‘year’)

phone_pop = phone.merge(population, how='inner', on=['Country','year'])
```

```python
#Print out the merge dataset

phone_pop
```

```python
#In the ‘Geo country codes’ rename the columns{‘country to ‘geo’, and ‘name’ to ‘Country’}

geo_code.rename(columns={"country": "geo", "name": "Country"}, inplace=True)
```

```python
#Print out the columns

geo_code.columns
```

```python
#select the ‘geo’ and ‘Country’ columns from the ‘geo_country’ dataframe

geo_country = geo_code[['geo','Country']]
```

```python
#Print out the selected columns

geo_country
```

```python
#join with your ‘merged population and cell phone’ data.

dataset = phone_pop.merge(geo_country, how='inner', on=['Country'])
```

```python
#Print out the new complete dataset

dataset
```

```python
#Calculate the number of cell phones per person, and add this column onto your dataframe. 
#(Hint: Might have to divide the total number of cell phones by the total number of population)

cell = dataset['cell_phones_total']
person = dataset['population']
cell_phones_per_person = cell / person
cell_phones_per_person
```

```python
# add this column onto your dataframe.
dataset['cell_phones_per_person'] = cell_phones_per_person
```

```python
# Modify the ‘geo’ column to make the ‘geocodes' uppercase instead of lowercase.
dataset['geo'] = dataset['geo'].str.upper() 
```

```python
dataset
```

```python
#Use the ‘.describe()’ method to Generate descriptive statistics and describe your dataframe's numeric columns

dataset.describe()
```

```python
#Use the ‘.describe()’ method to Generate descriptive statistics and describe your dataframe's non-numeric columns.
dataset.describe(include='all')
```

```python
#In 2017, what were the top 5 countries with the most cell phones total?

dataset_2017 = dataset[dataset['year'] == 2017]

#nlargest allows you to select the top values in a column
dataset_2017.nlargest(5,'cell_phones_total')
```
